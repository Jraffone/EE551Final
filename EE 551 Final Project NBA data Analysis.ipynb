```python
import pandas as pd
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns
sample_data = pd.read_excel('NBAplayerstats.xlsx')
```

```python
sample_data
```

```python
pd.set_option("display.precision",3)
sample_data
```

```python
Lebron = sample_data[sample_data.FULLNAME == 'LeBron James']
Lebron
```

```python
Luka = sample_data[sample_data.FULLNAME == 'Luka Doncic']
Luka
```

```python
Giannis = sample_data[sample_data.FULLNAME == 'Giannis Antetokounmpo']
Giannis
```

```python
plt.bar(Lebron.FULLNAME, Lebron.MPG)
plt.bar(Giannis.FULLNAME, Giannis.MPG)
plt.bar(Luka.FULLNAME, Luka.MPG)
plt.title("Minutes played per game for top 3 2019-2020 MVP candidates")
plt.ylabel("MPG")
plt.show()
```

```python
plt.bar(Lebron.FULLNAME, Lebron.PPG)
plt.bar(Giannis.FULLNAME, Giannis.PPG)
plt.bar(Luka.FULLNAME, Luka.PPG)
plt.title("Points Scored per game for top 3 2019-2020 MVP candidates")
plt.ylabel("PPG")
plt.show()
```

```python
plt.bar(Lebron.FULLNAME, Lebron.RPG)
plt.bar(Giannis.FULLNAME, Giannis.RPG)
plt.bar(Luka.FULLNAME, Luka.RPG)
plt.title("Rebounds per game for top 3 2019-2020 MVP candidates")
plt.ylabel("RPG")
plt.show()
```

```python
plt.bar(Lebron.FULLNAME, Lebron.APG)
plt.bar(Giannis.FULLNAME, Giannis.APG)
plt.bar(Luka.FULLNAME, Luka.APG)
plt.title("Assists per game for top 3 2019-2020 MVP candidates")
plt.ylabel("APG")
plt.show()
```

```python
plt.bar(Lebron.FULLNAME, Lebron.SPG)
plt.bar(Giannis.FULLNAME, Giannis.SPG)
plt.bar(Luka.FULLNAME, Luka.SPG)
plt.title("Steals per game for top 3 2019-2020 MVP candidates")
plt.ylabel("SPG")
plt.show()
```

```python
plt.bar(Lebron.FULLNAME, Lebron.BPG)
plt.bar(Giannis.FULLNAME, Giannis.BPG)
plt.bar(Luka.FULLNAME, Luka.BPG)
plt.title("Blocks per game for top 3 2019-2020 MVP candidates")
plt.ylabel("BPG")
plt.show()
```

```python
league_data = pd.read_excel('NBALeagueStats.xlsx')
league_data
```

```python
plt.bar("Season","PTS", data = league_data)
plt.xlabel("Season")
plt.xticks (rotation = 90)
plt.ylabel("Points")
plt.title("Average Points per game since 1990")
plt.show()
```

```python
plt.bar("Season","3PA", data = league_data)
plt.xlabel("Season")
plt.xticks (rotation = 90)
plt.ylabel("3PA")
plt.title("Average 3 point shots attempted per game since 1990")
plt.show()
```
